```python
#Q1.Write a python program to display all the header tags from wikipedia.org.

page= requests.get("https://en.wikipedia.org/wiki/Main_Page")
page
soup= BeautifulSoup(page.content)
soup
header_tags = []
for header in soup.find_all(["h1","h2","h3","h4","h5","h6"]):
    header_tags.append(header.name+" "+header.text.strip())
    
header_tags
```

```python
# Q2. Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame.

page1= requests.get("https://www.imdb.com/list/ls091520106/")
page1
soup1=BeautifulSoup(page1.content)
soup1
name = soup1.find_all("h3", class_="lister-item-header")
movies_name = [] 
for i in name:
    for j in i.find_all("a"):
        movies_name.append(j.text)


year = soup1.find_all("span",class_="lister-item-year text-muted unbold")
year_of_release = [] 
for k in year:
    a=k.text.replace('(','')
    year_of_release.append(a.replace(')','')) 


rating = soup1.find_all("div",class_="ipl-rating-star small")

IMDB_rating = [] 
for i in rating:
      IMDB_rating.append(float(i.text))
IMDB_top_100=pd.DataFrame({})
IMDB_top_100['movies_name']=movies_name
IMDB_top_100['year_of_release']=year_of_release
IMDB_top_100['IMDB_rating']=IMDB_rating  
IMDB_top_100
```

# Q5. Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:
#a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.
#b) Top 10 ODI Batsmen along with the records of their team and rating.
#c) Top 10 ODI bowlers along with the records of their team and rating.



```python
#a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

page4= requests.get("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")
page4
soup4=BeautifulSoup(page4.content)
soup4
team = soup4.find_all("span",class_='u-hide-phablet')
team_name = []
for i in team:
    team_name.append(i.text)
matches = []
for i in soup4.find_all("td",class_='rankings-block__banner--matches'): 
    matches.append(i.text)
    
points = []
for i in soup4.find_all("td",class_='rankings-block__banner--points'):
    points.append(i.text)
    
ratings = []
for i in soup4.find_all("td",class_='rankings-block__banner--rating u-text-right'):
    ratings.append(i.text.replace("\n",""))
    
new_list = []
for i in soup4.find_all("td",class_='table-body__cell u-center-text'):
    new_list.append(i.text)
for i in range(0,len(new_list)-1,2):
    matches.append(new_list[i]) 
    points.append(new_list[i+1]) 
for i in soup4.find_all("td",class_='table-body__cell u-text-right rating'):
    ratings.append(i.text)
import pandas as pd
df1=pd.DataFrame({})
df1['Team']=team[:10]
df1['Matches']=matches[:10]
df1['Points']=points[:10]
df1['Ratings']=ratings[:10]
df1
```

```python
#b) Top 10 ODI Batsmen along with the records of their team and rating.
page5=requests.get("https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting")
page5
soup5=BeautifulSoup(page5.content)
soup5
players=[]
for i in soup5.find_all('div',class_="rankings-block__banner--name-large"):
    players.append(i.text)

team_name=[]
for i in soup5.find_all('div',class_="rankings-block__banner--nationality"):
    team_name.append(i.text.replace("/n",""))
    
ratings=[]
for i in soup5.find_all('div',class_="rankings-block__banner--rating"):
    ratings.append(i.text)
    
for i in soup5.find_all('td',class_="table-body__cell rankings-table__name name"):
    for j in i.find_all('a'):
        players.append(j.text)
        
for i in soup5.find_all('span',class_="table-body__logo-text"):
    team_name.append(i.text.replace("\n",""))
    
for i in soup5.find_all('td',class_="table-body__cell rating"):
    ratings.append(i.text)
    
df3=pd.DataFrame({})
df3['Players']=players[:10]
df3['Team_Name']=team_name[:10]
df3['Ratings']=ratings[:10]
df3
```

```python
#c) Top 10 ODI bowlers along with the records of their team and rating.
page6= requests.get("https://www.icc-cricket.com/rankings/mens/player-rankings/odi/bowling")
page6
soup6=BeautifulSoup(page6.content)
soup6
players = []
for i in soup6.find_all("div",class_='rankings-block__banner--name-large'):
    players.append(i.text)

team_name = []
for i in soup6.find_all("div",class_='rankings-block__banner--nationality'):
    team_name.append(i.text.replace("\n",""))

rating = []
for i in soup6.find_all("div",class_='rankings-block__banner--rating'): 
    rating.append(i.text)
    
for i in soup6.find_all("td",class_='table-body__cell rankings-table__name name'):
    for j in i.find_all('a'):
        players.append(j.text)
for i in soup6.find_all("span",class_='table-body__logo-text'):
    team_name.append(i.text)
for i in soup6.find_all("td",class_='table-body__cell rating'):
    rating.append(i.text)

df4=pd.DataFrame({})
df4['Player']=players[:10]
df4['Team']=team_name[:10]
df4['Rating']=rating[:10]
df4
```

Q7Write a python program to scrape details of all the posts from coreyms.com. Scrape the heading, date, content
and the code for the video from the link for the youtube video from the post

```python
url = "https://coreyms.com/"
page15 = requests.get(url)

soup15 = BeautifulSoup(page15.content)



articles = soup15.find_all('article')

data = []
for article in articles:
    header = article.find("header")
    
    
    title = header.find("h2").text
    
    date = header.find("time", class_="entry-time").text
    div_block = article.find("div", class_="entry-content")
    
    paragraph_content = div_block.find("p").text
    try:
        youtube_link = div_block.find("iframe", class_="youtube-player")['src']
        video_code = youtube_link.split("?")[0].split("/")[-1]
    except:
        video_code = "NA"
    
    data.append([title,date, paragraph_content, video_code])
    print(title)
    
```

```python
columns = ["Title","Date", "Content","Video Code"]
df15 = pd.DataFrame(data, columns=columns)
df15
```

9) Write a python program to scrape mentioned details from dineout.co.in :
i) Restaurant name
ii) Cuisine
iii) Location
iv) Ratings
v) Image URL

```python
page16= requests.get("https://www.dineout.co.in/delhi-restaurants/buffet-special")
page16
```

```python
soup16=BeautifulSoup(page16.content)
soup16
```

```python
name=[]
for i in soup16.find_all("div", class_="restnt-info cursor"):
    name.append(i.text)
location=[]
for i in soup16.find_all("div", class_="restnt-loc ellipsis"):
    location.append(i.text)

price = []
cuisine = []
for i in soup16.find_all("span", class_="double-line-ellipsis"):
    price.append(i.text.split('|')[0])
    cuisine.append(i.text.split('|')[1])

rating=[]
for i in soup16.find_all("div", class_="restnt-rating rating-3"):
    rating.append(i.text)
for i in soup16.find_all("div", class_="restnt-rating rating-4"):
    rating.append(i.text)

images = []
for i in soup16.find_all("img", class_="no-img"):
    images.append(i['data-src'])
```

```python
print(len(name), len(location), len(price), len(cuisine), len(rating), len(images))
```

```python
df16=pd.DataFrame({})
df16['Restaurant Name']=name
df16['Location']=location
df16['Price']=price 
df16['Cuisine']=cuisine  
df16['Rating']=rating  
df16['IMAGES']=images
df16
```
